# ADMINISTRACIÓN DE AGENDA DE CONTACTOS
# -------------------------------------------------

```python
def mostrar_menu():
    print("\n" + "═"*30)
    print("      MENÚ PRINCIPAL")
    print("" + "═"*30)
    print("1. Agregar contacto")
    print("2. Listar contactos")
    print("3. Buscar contacto")
    print("4. Editar contacto")
    print("5. Eliminar contacto")
    print("6. Salir")
    print("═"*30)
```

```python
def pedir_datos_contacto():
    print("\n(Escriba 'salir' en cualquier momento para cancelar)")
    print(" ")
    # Validación de Nombre
    while True:
        nombre = input("Nombre: ").strip()
        if nombre.lower() == 'salir': return None
        if nombre: break
        print("⚠️ El nombre no puede estar vacío.")

    # Validación de Teléfono (Solo números)
    while True:
        telefono = input("Teléfono (solo números): ").strip()
        if telefono.lower() == 'salir': return None
        if telefono.isdigit(): break
        print("⚠️ Error: Ingrese solo dígitos numéricos.")

    # Validación de Correo
    while True:
        correo = input("Correo electrónico: ").strip()
        if correo.lower() == 'salir': return None
        if "@" in correo and "." in correo: break
        print("⚠️ Error: Formato de correo inválido (ej: nombre@mail.com).")

    # Dirección
    direccion = input("Dirección: ").strip()
    if direccion.lower() == 'salir': return None

    return {
        "nombre": nombre,
        "telefono": telefono,
        "correo": correo,
        "direccion": direccion
    }
```

```python
def agregar_contacto(lista):
    print("\n--- AGREGAR NUEVO CONTACTO ---")
    nuevo = pedir_datos_contacto()
    if nuevo:
        lista.append(nuevo)
        print("✅ Contacto guardado con éxito.")
    else:
        print("❌ Operación cancelada.")
```

```python
def listar_contactos(lista):
    print("\n--- LISTADO DE CONTACTOS ---")
    if not lista:
        print("📭 La agenda está vacía.")
        return False

    for i, c in enumerate(lista):
        print(f"[{i}] {c['nombre']} | Tel: {c['telefono']} | {c['correo']} | {c['direccion']}")
    return True
```

```python
def buscar_contacto(lista_contactos):
    if not lista_contactos:
        print("\n📭 No hay contactos para buscar.")
        return

    termino = input("\nIngrese nombre o teléfono a buscar: ").lower()
    encontrados = False

    print("\n--- RESULTADOS DE BÚSQUEDA ---")
    for c in lista_contactos:
        # Buscamos coincidencias en nombre o teléfono
        if termino in c['nombre'].lower() or termino in c['telefono']:
            print(f"• {c['nombre']} | Tel: {c['telefono']} | Correo: {c['correo']}")
            encontrados = True

    if not encontrados:
        print("🔎 No se encontraron coincidencias.")
```

```python
def editar_contacto(lista):
    print("\n--- EDITAR CONTACTO ---")
    if not listar_contactos(lista): return

    idx_txt = input("\nIngrese el número (índice) a editar: ").strip()

    if idx_txt.isdigit():
        idx = int(idx_txt)
        if 0 <= idx < len(lista):
            print(f"Editando a: {lista[idx]['nombre']}")
            nuevo = pedir_datos_contacto()
            if nuevo:
                lista[idx] = nuevo
                print("✅ Contacto actualizado.")
            else:
                print("❌ Edición cancelada.")
        else:
            print("⚠️ El índice no existe.")
    else:
        print("⚠️ Debe ingresar un número válido.")
```

```python
def eliminar_contacto(lista):
    print("\n--- ELIMINAR CONTACTO ---")
    if not listar_contactos(lista): return

    idx_txt = input("\nIngrese el número (índice) a eliminar: ").strip()

    if idx_txt.isdigit():
        idx = int(idx_txt)
        if 0 <= idx < len(lista):
            eliminado = lista.pop(idx)
            print(f"🗑️ Contacto '{eliminado['nombre']}' eliminado.")
        else:
            print("⚠️ El índice no existe.")
    else:
        print("⚠️ Entrada inválida.")
```

```python
def main():
    agenda = []

    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción: ").strip()

        if opcion == "1":
            agregar_contacto(agenda)
        elif opcion == "2":
            listar_contactos(agenda)
        elif opcion == "3":
            buscar_contacto(agenda)
        elif opcion == "4":
            editar_contacto(agenda)
        elif opcion == "5":
            eliminar_contacto(agenda)
        elif opcion == "6":
            print("\n Saliendo de la agenda.")
            break
        else:
            print("⚠️ Opción no válida.")

if __name__ == "__main__":
    main()
```
